```python
x = dfnew["longitude"].to_numpy()
y = dfnew["latitude"].to_numpy()
z = dfnew[header[-100]]-dfnew[header[1000]]

from matplotlib.colors import LinearSegmentedColormap
norm=plt.Normalize(-2,2)
cmap = LinearSegmentedColormap.from_list("", ["green","yellow","red"])

plt.figure(figsize=(20,8))
plt.scatter(x,y, c=z,cmap=cmap,s=20,vmin = 10, vmax = 10)
#plt.scatter(x,y, c=z,cmap=cmap,s=20,vmin = 20, vmax = 32)
plt.colorbar().ax.set_ylabel("°C",rotation=0)
plt.xlim([105,114.5])
plt.ylim([-9,-5])
plt.title("Temperatur Pulau Jawa")
plt.tick_params(left = False, right = False , labelleft = False , 
                labelbottom = False, bottom = False) 
grid_X,grid_Y = np.meshgrid(np.linspace(105,115,200),np.linspace(-9,-5,200))
plt.scatter(grid_X.reshape(-1),grid_Y.reshape(-1))
plt.show()
```

```python
# Example data
data1 = np.array([x,y]).T  
data2 = np.array([grid_X.reshape(-1),grid_Y.reshape(-1)]).T
threshold = 0.035  # Set your threshold distance

# Function to calculate distances
def exclude_close_points(data1, data2, threshold):
    # Calculate the squared distance matrix between data1 and data2
    distances = np.sqrt(((data2[:, None, :] - data1[None, :, :]) ** 2).sum(axis=2))
    # Check if any distances are below the threshold
    mask = np.all(distances > threshold, axis=1)
    # Filter data2 points based on the mask
    return data2[mask]

# Exclude close points
filtered_data2 = exclude_close_points(data1, data2, threshold)
plt.figure(figsize=(20,7))
plt.scatter(x,y, c=z,cmap=cmap,s=20,vmin = 20, vmax = 32)
plt.xlim([105,115])
plt.ylim([-9,-5.5])
plt.title("Temperatur Pulau Jawa (Hijau: Titik Observasi Data)")
plt.tick_params(left = False, right = False , labelleft = False , 
                labelbottom = False, bottom = False) 
plt.scatter(filtered_data2[:,0],filtered_data2[:,1])
plt.show()
```
