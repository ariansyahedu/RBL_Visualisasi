```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

path = "../data/"
ffp = []
for root, dirs, files in os.walk(path):
    for file in files:
        if file.endswith(".csv"):
            ffp.append(os.path.join(root, file)) 

for i in ffp:
    print(i)
```

```python
from datetime import datetime, timedelta

start_date = datetime(2011, 1, 1)  # Starting date: January 1, 2011

# Generate list of all dates
header = [(start_date + timedelta(days=i)).strftime('%Y-%m-%d') for i in range(4800)]
header.insert(0,"latitude")
header.insert(0,"longitude")
dfc = pd.read_csv(ffp[1], names = header)
```

```python
x = dfc["longitude"].to_numpy()
y = dfc["latitude"].to_numpy()
```

```python
plt.figure(figsize=(20,8))
plt.scatter(x,y, c=dfc[header[1000]],cmap="coolwarm",s=20)
plt.colorbar().ax.set_ylabel("°C",rotation=0)
plt.xlim([105,114.5])
plt.ylim([-9,-5])
plt.title("Temperatur Pulau Jawa")
plt.tick_params(left = False, right = False , labelleft = False , 
                labelbottom = False, bottom = False) 
plt.show()
```
